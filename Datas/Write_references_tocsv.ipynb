```python
import os
import pandas as pd
import re
```

```python
data=list()
directory = os.getcwd()+"/Oss2"
for file in os.listdir(directory):
    if file.endswith('.mp3'):
        temp =["2"]+re.split("_|.mp3",file)[1:-1]+["mp3"]+[file]
        data.append(temp)
        
data.sort(key=lambda x: x[1])
data
```

```python
df = pd.DataFrame(data=data,columns=["Film","Scene","Reference","Format",'filename'])
df
```

```python
df.to_csv(os.getcwd()+"references2.csv")
```
